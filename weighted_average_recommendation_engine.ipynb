```python
import pandas as pd
import numpy as np
```

```python
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
```

```python
credits_column_renamed = credits.rename(index = str,columns = {"movie_id":"id"})
```

```python
credits_column_renamed.head()
```

```python
movies_df_merge = movies.merge(credits_column_renamed,on="id")
```

```python
movies_df_merge.head()
```

```python
movies_df_merge.shape
```

```python
movies_df_merge = movies_df_merge.drop(columns = ["homepage","title_x","title_y","status","production_countries"])
```

```python
movies_df_merge.head()
```

```python
movies_df_merge.shape
```

```python
movies_df_merge.info()
```

## W = (Rv + Cm)/v+m

### W = Weighted Rating
### v = number of votes for the movie
### R = Average for the movie a number b/w 0 to 10
### m = min votes required to be listed in the top 250
### C = Mean vote accross the whole rating

```python
v = movies_df_merge['vote_count']
```

```python
R = movies_df_merge["vote_average"]
```

```python
C = movies_df_merge["vote_average"].mean()
print(C)
```

```python
m = movies_df_merge["vote_count"].quantile(0.70)
print(m)
```

```python
movies_df_merge["weighted_rating"] = ((R*v) + (C*m))/(v+m)
```

```python
movies_df_merge.head()
```

```python
movies_sorted_rankings = movies_df_merge.sort_values('weighted_rating',ascending = False)
movies_sorted_rankings[['original_title','vote_count','vote_average','weighted_rating','popularity','runtime','release_date']][:20]
```

### 50% importance to weighted rating and 50% importance to popularity

```python
from sklearn.preprocessing import MinMaxScaler

scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(movies_df_merge[['weighted_rating','popularity']])
movie_normalized_df = pd.DataFrame(movie_scaled_df,columns = ['weighted_rating','popularity'])
movie_normalized_df.head()
```

```python
movies_df_merge[['normalized_weighted_rating','normalized_popularity']] = movie_normalized_df
```

```python
movies_df_merge.head()
```

```python
movies_df_merge['score'] = movies_df_merge['normalized_weighted_rating']*0.5 + movies_df_merge['normalized_popularity'] * 0.5
movies_df_merge.head()
```

```python
movies_scored_df = movies_df_merge.sort_values(['score'],ascending = False)
```
